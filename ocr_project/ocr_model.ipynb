# 2. OCR로 이미지에서 텍스트 추출하기

````python
from paddleocr import PaddleOCR
ocr = PaddleOCR(
    use_doc_orientation_classify = False,
    use_doc_unwarping = False,
    use_textline_orientation = False
)

img_path = "images/webtoon1.jpeg"
results = ocr.predict(input = img_path)
results
````

# 3. OCR 활용 1

````python
from PIL import Image
import matplotlib.pyplot as plt

img_path = "images/webtoon1.jpeg"
image = Image.open(img_path)

plt.figure(figsize=(5,5))
plt.imshow(image)
plt.axis("off")
plt.show()
````

````python
# 이미지에서 박스 위치를 찾아 그 말풍선에 있는 텍스트를 없애고 새로운 텍스트를 넣어보자
````

````python
from PIL import ImageDraw, ImageFont

# 그리기 준비
image_new = image.copy()
draw = ImageDraw.Draw(image_new)

# 텍스트 쓸 준비
font_path = "C:/Windows/Fonts/malgun.ttf"
font = ImageFont.truetype(font_path, 25)
````

````python
# Box 생김새 확인 (x1, y1, x2, y2) -- 좌측상단, 우측하단
result["rec_boxes"][0].tolist()
````

````python
result["rec_texts"]
````

````python
my_text_list=["오늘 날씨 진짜 좋다", "오늘 뭐 먹지", "도망가고 싶다"]
````

````python
for idx, (x1, y1, x2, y2) in enumerate(result["rec_boxes"]):
    print(x1, y1, x2, y2)
    draw.rectangle([x1, y1, x2, y2], outline="white", fill="green")
    draw.text([x1, y1, x2, y2], my_text_list[idx], font=font, fill="black")
````
