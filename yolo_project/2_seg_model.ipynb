# YOLO-seg 모델

# 1. YOLO11n-seg 모델 사용하기

## 1) 모델 불러오기

````python
from ultralytics import YOLO

model = YOLO("model/yolo11n-seg.pt")
model
````

## 2) 이미지 예측하기

````python
img_url = "https://ultralytics.com/images/bus.jpg"

results = model(img_url, save=True, project="myresult")
````

````python
print(len(results))
````

````python
result = results[0]
result
````

````python
names = result.names
names
````

````python
boxes = result.boxes
boxes
````

````python
masks = result.masks
print(len(masks.data))  # masks 에는 6개의 이미지가 들어있다
````

````python
# 이미지 출력하기
import matplotlib.pyplot as plt

plt.imshow(result.masks.data[1])
plt.axis("off")
plt.show()
````

# 2. FastSAM 모델 사용하기

## 1) 모델 불러오기

````python
from ultralytics import FastSAM

model = FastSAM("model/FastSAM-s.pt")
model
````

## 2) 이미지 예측하기

````python
source = "images/dog.jpeg"

everything_results = model(source, save=True, project="myresult", retina_masks=True, conf=0.5, iou=0.9)
len(everything_results)

# retina_masks : 더 정교하게 해달라
````

````python
result = everything_results[0]
result
````

````python
masks = result.masks
masks
````

````python
print(len(masks))
````

````python
# 이미지 출력하기
import matplotlib.pyplot as plt

plt.imshow(result.masks.data[16])
plt.axis("off")
plt.show()
````

## 3) 이미지와 텍스트로 예측하기

````python
results = model(source, texts="a photo of a cap", save=True, project="myresult")
````
