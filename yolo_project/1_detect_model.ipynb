# 1. 모델 불러오기

````python
from ultralytics import YOLO

# 모델 불러오기
model = YOLO("model/yolo11n.pt")  # 다운로드 위치
model
````

# 2. 이미지 예측

````python
results = model("./images/yolo_image.jpg", save=True, project="myresult")   # model(), model.predict()
print(type(results))
print(len(results)) # 예측한 이미지 수
````

````python
result = results[0] # 이미지 하나에 대한 결과물
result
````

````python
# result에는 이미지에 대한 예측 결과 또는 정보가 담겨있음
names = result.names    # yolo11n 모델이 찾을 수 있는 객체 딕셔너리
boxes = result.boxes    # 이미지에서 찾은 객체 정보들이 들어있다
print(names)
print(boxes)
````

````python
print(type(boxes))
print(boxes.cls)    # 예측결과(class) : 0-person, 1-bicycle, 25-umbrella
print(boxes.conf)   # 신뢰도(confidence)
print(boxes.data)   # x1, y1, x2, y2, conf, cls
# boxes 길이 : 찾은 객체의 수
````

````python
for idx in boxes.cls:
    # print(idx, int(idx))
    print(f"{names[int(idx)]}를 찾았습니다!")
````

````python
for x1, y1, x2, y2, conf, cls in boxes.data:  # boxes.data(x1, y1, x2, y2, conf, cls)
    print(x1, y1, x2, y2, conf, cls)
````

````python
# image 하나로 예측하기
# results --- 이미지 개수만큼 리스트로 담겨 있었다 len(results) -> 1
# results = results[0]
# results.names --- 딕셔너리로 탐지할 수 있는 객체 저장되어 있음
# boxes = result.boxes --- 객체 인식한 박스 정보들이 들어가 있음
# boxes.cls, boxes.conf, boxes.data
# cls : 객체 예측한 결과
# conf : 신뢰도
# data : 박스좌표(x1, y1, x2, y2), 신뢰도(conf), 결과(cls)
````
